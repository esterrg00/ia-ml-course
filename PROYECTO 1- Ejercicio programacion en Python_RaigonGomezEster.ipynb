## PARTE 1

````python
import pandas as pd

def puntos(granPremio, clas_actual):
    dicPuntos ={}
    clasif_data = pd.read_csv(clas_actual, sep=',')
    dic_clasif = clasif_data.to_dict()
    listaPts = list(dic_clasif['Puntos'].items())
    for pos,piloto in list(granPremio['posiciones'].items()):      
        for i,nombrePiloto in list(dic_clasif['nombre'].items()):
            if piloto == nombrePiloto.strip():
                if pos == '1':
                    dicPuntos[piloto] = listaPts[i][1] + 25
                elif pos == '2':
                    dicPuntos[piloto] = listaPts[i][1] +20
                elif pos == '3':
                    dicPuntos[piloto] = listaPts[i][1] +16
                elif pos == '4':
                    dicPuntos[piloto] = listaPts[i][1] +13
                elif pos == '5':
                    dicPuntos[piloto] = listaPts[i][1] +11
                elif pos == '6':
                    dicPuntos[piloto] = listaPts[i][1] +10
                elif pos == '7':
                    dicPuntos[piloto] = listaPts[i][1] +9
                elif pos == '8':
                    dicPuntos[piloto] = listaPts[i][1] +8
                elif pos == '9':
                    dicPuntos[piloto] = listaPts[i][1] +7
                elif pos == '10':
                    dicPuntos[piloto] = listaPts[i][1] +6
                elif pos == '11':
                    dicPuntos[piloto] = listaPts[i][1] +5
                elif pos == '12':
                    dicPuntos[piloto] = listaPts[i][1] +4
                elif pos == '13':
                    dicPuntos[piloto] = listaPts[i][1] +3
                elif pos == '14':
                    dicPuntos[piloto] = listaPts[i][1] +2
                elif pos =='15':
                    dicPuntos[piloto] = listaPts[i][1] +1
                                        
    return dicPuntos
````

## PARTE 2 

División del problema en dos etapas:

**2.1** Actualizar columna puntos del DataFrame a partir del csv

**2.2** Reordenar DataFrame según la nueva puntuación 

````python
nuevaPuntuacion = puntos(resultados, 'Clasificacion-actual.csv')
````

### 2.1 Actualizar la variable puntos a partir del csv

````python
def updateScores(newPuntuacion, clas_actual):
        
    df = pd.read_csv(clas_actual, sep=',')

    #posicionesActual = list(df.to_dict().items())[0][1]
    ptsActualizados = df[['nombre','Puntos']]
    lst =ptsActualizados.values.tolist()

    #strippeamos los nombres de los pilotos del CSV original para poder comparar las cadenas:
    for linea in range(len(lst)):
        lst[linea][0]=lst[linea][0].strip()
        
    # Actualizamos la lista del dataFrame con las nuevas puntuaciones calculadas en PARTE 1
    for item in range(len(lst)):
        piloto = lst[item][0]
        for nombreP, pts in list(newPuntuacion.items()):
            if piloto == nombreP:
                lst[item][1] = pts
                
    return lst
````

````python
puntosUpdated = updateScores(nuevaPuntuacion,'Clasificacion-actual.csv')
puntosUpdated [:4]
````

### 2.2 Reordenar clasificación según nueva puntación

````python
def reorderClass (lstPuntosUpdated, clas_actual):
    df = pd.read_csv(clas_actual, sep=',')
    
    # Se pasa la lista de las nuevas puntuaciones a DataFrame
    puntosDF = pd.DataFrame(lstPuntosUpdated,columns=['name_piloto','puntosUpdated'])
    
    # Sustituimos las puntuaciones del DataFrame original con las nuevas puntuaciones
    df['Puntos'] = puntosDF['puntosUpdated']

    # Reordenamos la clasificación según la nueva puntuación
    df = df.sort_values(by='Puntos',ascending=False).reset_index(drop=True)
    
    for index in range(df.size):
        df['Pos.'][index] = index+1
        
    return df
````

````python
clas_updated = reorderClass(puntosUpdated,'Clasificacion-actual.csv')

clas_updated.head()
````

## PARTE 3

````python
def updateClasXGP (results, ficheroGP, ficheroTodosGP):
    
    nuevosPuntos = puntos(results, ficheroGP)
    
    lstPuntosUpdated = updateScores(nuevosPuntos, ficheroGP)
    
    dataFrameGP = reorderClass(lstPuntosUpdated, ficheroGP)
    
    df = pd.read_csv(ficheroTodosGP, sep=',')
    
    lugar = results['gran_premio']
    
    #Comprobamos si contiene el lugar del GranPremio está en el dataFrame, si no pues se crea una nueva columna para rellenarla
    if lugar not in list(df.columns):
        df.insert(lugar)
    
    puntosGP = df[['Nombre','Puntuación', lugar]]     
    lstPuntosGP = puntosGP.values.tolist()
    
    clasPuntosGP = dataFrameGP[['nombre','Puntos']]
    lstClasPuntosGP = clasPuntosGP.values.tolist()
    
    for participante in range(len(lstPuntosGP)):
        piloto = lstPuntosGP[participante][0]
        for datosPiloto in range(len(lstClasPuntosGP)):
            pilotoName = lstClasPuntosGP[datosPiloto][0]
            if piloto == pilotoName:
                
                if type(lstPuntosGP[participante][2]) == str:
                    lstPuntosGP[participante][2] = lstClasPuntosGP[datosPiloto][1] - lstPuntosGP[participante][1]
                else:
                    lstPuntosGP[participante][2] = lstPuntosGP[participante][2] + (lstClasPuntosGP[datosPiloto][1] 
                                                                                   - lstPuntosGP[participante][1])
                
                lstPuntosGP[participante][1] = lstClasPuntosGP[datosPiloto][1]
    
    return {lugar: lstPuntosGP}
````

````python
datosTodosGP = updateClasXGP(resultados,'Clasificacion-actual.csv','Clasificaciones por cada gran premio.csv')
datosTodosGP
````
